```python
import random

#1から100までの整数の中から、ランダムな整数を代入する
var = random.randint(1, 100)

#実行した時にどの数字が変数varとして選ばれたのか出力
print(var)

#3と5の両方で割り切れる(15の倍数)か確認
if var % 3 == 0 and var % 5 == 0 :
    print("FizzBuzz") #15の倍数なら「FizzBuzz」を出力

#上記に当てはまらず、3で割り切れる(3の倍数)か確認
elif var % 3 == 0 :
    print("Fizz") #3の倍数なら「Fizz」を出力

#上記に当てはまらず、5で割り切れる(5の倍数)か確認
elif var % 5 == 0 :
    print("Buzz") #5の倍数なら「Buzz]を出力

#どの条件にも当てはまらない場合
else :
    print(var) #選ばれた数字をそのまま出力
```
