```python
array = ["水", "金", "地", "火", "木", "土", "天", "海", "冥"]
#for文による繰り返し処理
#リストの中身を順番にplanetという変数に入れます
for planet in array:
    print(planet)

#while文による繰り返し処理
#index(何番目か)を0からスタートさせる
index = 0

#今回は中身は9個なので、9未満の間繰り返す
while index < 9:
    print(array[index])

    #indexを1ずつ増やして次の繰り返しへ
    index = index + 1
```
