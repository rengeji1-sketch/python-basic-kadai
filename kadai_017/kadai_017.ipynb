```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def check_adult(self):  #Humanクラスの中に年齢チェックする関数を入れる
        if self.age >= 20:
            print(f"{self.name}は大人です。")
        else:  #20以上ではない場合はfalseになる
            print(f"{self.name}は大人ではありません。")
    #Humanクラスのインスタンスを複数生成
human1 = Human("一子", 25)
human2 = Human("二子", 15)
human3 = Human("三郎", 30)
human4 = Human("四郎", 18)
  #生成したインスタンスをリストに追加
humans_list = [human1, human2, human3, human4]
print("大人かどうかチェックします。")  #ここまでのコードが読み込まれているかチェック
  #リストからhumanの数だけ繰り返す
for human in humans_list:
      #取り出したhumanを使ってcheck_adult関数を使う
    human.check_adult()
print("チェック終了しました。")  #チェック終了の報告
```
