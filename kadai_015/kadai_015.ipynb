```python
class Human: #Humanクラスを作成
    #Humanクラスで使う関数を定義する
    def __init__(self, name, age):  #クラスの中で最初からname, ageは持っていることにする
        self.name = name
        self.age = age
    #Humanクラスで使うprintinfoという関数を定義する
    def printinfo(self):
        print(f"こんにちは。私は{self.name}です。年齢は{self.age}です")  #この関数の中で、printを使って出力まで行う

human1 = Human("桜", 3)  #出力するためにHumanクラスの新しいインスタンスを作成し、human1という変数に代入する

human1.printinfo()  #printinfoの関数の中ですでに出力もするように定義されているので、ここではprintを使わず、human1の情報を持つprintinfoを呼び出すだけでいい。
```
